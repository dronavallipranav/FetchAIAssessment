```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data_daily.csv")
```

```python
df["month"] = df["# Date"].str[5:7]
```

```python
df
```

```python
df_plot = df.groupby('month').agg(total_receipts=('Receipt_Count', 'sum'), average_receipts=('Receipt_Count', 'mean')).reset_index()
```

```python
df_plot
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

ax[0].bar(df_plot['month'], df_plot['total_receipts'], color='skyblue')
ax[0].set_xlabel('Month')
ax[0].set_ylabel('Total Receipts')
ax[0].set_title('Sum of Receipts')

ax[1].bar(df_plot['month'], df_plot['average_receipts'], color='salmon')
ax[1].set_xlabel('Month')
ax[1].set_ylabel('Average Receipts')
ax[1].set_title('Average Recepits')

plt.tight_layout()
plt.show()
#Seems like the receipt_count is montonically increasing
#My assumption is that company growth plays a much bigger role than seasonality
#It might still be a good idea to add some feature cols like Day/Month to catch possible trends in the data
```
